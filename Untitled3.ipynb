# **Deteccin de Outliers**



```python
outliers_iqr = {}

for col in numericas:
    Q1 = df_pricing[col].quantile(0.25)
    Q3 = df_pricing[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df_pricing[(df_pricing[col] < lower) | (df_pricing[col] > upper)]
    porcentaje_outliers = round(len(outliers) / len(df_pricing) * 100, 2)
    outliers_iqr[col] = porcentaje_outliers
    print(f"{col}: {porcentaje_outliers}% de outliers")
```

```python
outliers_z = {}

for col in numericas:

    z_scores = np.abs(stats.zscore(df_pricing[col], nan_policy='omit'))
    mask = (z_scores > 3)
    outliers = df_pricing[mask]
    porcentaje_outliers = round(mask.sum() / len(df_pricing) * 100, 2)
    outliers_z[col] = porcentaje_outliers

    print(f"{col}: {porcentaje_outliers}% de outliers (Z-Score > 3)")
```
